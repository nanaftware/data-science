```python
# use MinMaxScaler to scale data
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()

X = data_new.drop(['Date','btc_market_price'], axis =1)
y = data_new['btc_market_price']

#split data
from sklearn.model_selection import train_test_split
X_train_org, X_test_org, y_train, y_test = train_test_split(X,y, test_size = 0.2,random_state = 20)

X_train = scaler.fit_transform(X_train_org)
X_test = scaler.transform(X_test_org)
```

# Train data with validation

```python
from sklearn.model_selection import cross_val_score
scores = {}
cv_s= {}
```

# vecinos K

```python
from sklearn.neighbors import KNeighborsRegressor
training_accuracy = []
test_accuracy = []
neighbor_seetings = range(1,7)

for x in neighbor_seetings:
    knr = KNeighborsRegressor(n_neighbors = x)
    knr.fit(X_train, y_train)
    training_accuracy.append(knr.score(X_train, y_train))
    test_accuracy.append(knr.score(X_test,y_test))

# Show the relationship of accuarcy and n_neighbors
plt.plot(neighbor_seetings, training_accuracy, label = 'training accuracy')
plt.plot(neighbor_seetings, test_accuracy, label = 'test accuracy')
plt.ylabel("Accuracy")
plt.xlabel('n_neighbors')
plt.legend()
```

```python
# El mejor es el 3
scores['K-neighbors regression'] = [training_accuracy[2],test_accuracy[2]]
```

```python
knr = KNeighborsRegressor(n_neighbors = 3)
cv_scores = cross_val_score(knr, X_train, y_train,cv = 5)
print('Cross-validation training scores (5-fold):', cv_scores)
print('Mean cross-validation training score (5-fold): {:.3f}'.format(np.mean(cv_scores)))
cv_test_scores = cross_val_score(knr, X_test, y_test,cv = 5)
cv_s['K-neighbors regression']=[np.mean(cv_scores),np.mean(cv_test_scores)]
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()

lr.fit(X_train, y_train)
print('Train score: {}'.format(lr.score(X_train, y_train)))
print('Test score: {}'.format(lr.score(X_test,y_test)))
scores['LinearRegression']=[lr.score(X_train, y_train),lr.score(X_test,y_test)]
```

```python
cv_scores = cross_val_score(lr, X_train, y_train,cv = 5)
print('Cross-validation training scores (5-fold):', cv_scores)
print('Mean cross-validation training score (5-fold): {:.3f}'.format(np.mean(cv_scores)))
cv_test_scores = cross_val_score(lr, X_test, y_test,cv = 5)
cv_s['LinearRegression']=[np.mean(cv_scores),np.mean(cv_test_scores)]
```

```python
# Alpha by default is 1.0
from sklearn.linear_model import Ridge
ridge = Ridge()
ridge.fit(X_train, y_train)
print('Train score: {}'.format(ridge.score(X_train, y_train)))
print('Test score: {}'.format(ridge.score(X_test,y_test)))
```

```python
a =np.linspace(0.001,25,1000)
training_accuracy = []
test_accuracy = []
for x in a:
    ridge = Ridge(alpha = x)
    ridge.fit(X_train, y_train)
    training_accuracy.append(ridge.score(X_train, y_train))
    test_accuracy.append(ridge.score(X_test,y_test))

plt.plot(a, training_accuracy, label = 'training accuracy')
plt.plot(a, test_accuracy, label = 'test accuracy')
plt.ylabel("Scores")
plt.xlabel('alpha')
plt.legend()
```

```python
# Use GridSearchCV to find the best alpha
from sklearn.model_selection import GridSearchCV

ridge = Ridge()
param_grid = {'alpha':[0.001,0.01, 0.1,1]}

grid_search = GridSearchCV(ridge, param_grid, cv = 5)
grid_search.fit(X_train, y_train)

#use gridsearch to get the best parameters, alpha=0.001
print('Best parameters{}'.format(grid_search.best_params_))
print('Best score {:.2f}'.format(grid_search.best_score_))
```

```python
ridge = Ridge(alpha = 0.001)
ridge.fit(X_train, y_train)
scores['Ridge']=[ridge.score(X_train, y_train),ridge.score(X_test,y_test)]
```

```python
# Cross Validation
cv_scores = cross_val_score(ridge, X_train, y_train,cv = 5)
print('Cross-validation training scores (5-fold):', cv_scores)
print('Mean cross-validation training score (5-fold): {:.3f}'.format(np.mean(cv_scores)))
cv_test_scores = cross_val_score(ridge, X_test, y_test,cv = 5)
cv_s['Ridge']=[np.mean(cv_scores),np.mean(cv_test_scores)]
```

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from datetime import timedelta

# 1. Cargar los datos del archivo CSV
try:
    df=pd.read_csv("../input/bitcoin/BitCoin.csv")
except FileNotFoundError:
    print("Error: No se encontrÃ³ el archivo 'btcusd_1-min_data.csv'. Por favor, asegÃºrate de que el archivo estÃ© en el directorio correcto.")
    exit()
```

```python
# 2. Preprocesar los datos
# Convertir timestamp a datetime (asumiendo formato UNIX en segundos)
df['timestamp'] = pd.to_datetime(df['timestamp'], unit='s')
# Ordenar por fecha
df = df.sort_values('timestamp')
# Crear una columna de tiempo en minutos desde el inicio
df['minutes'] = (df['timestamp'] - df['timestamp'].min()).dt.total_seconds() / 60

# Variables para la regresiÃ³n
X = df['minutes'].values.reshape(-1, 1)  # Tiempo en minutos
y = df['close'].values  # Precio de cierre
```

```python
# 3. Ajustar la regresiÃ³n lineal
modelo = LinearRegression()
modelo.fit(X, y)
```

```python
# 4. ExtrapolaciÃ³n lineal (predecir 30 dÃ­as en el futuro)
# 30 dÃ­as = 30 * 24 * 60 minutos = 43200 minutos
ultimo_tiempo = X[-1][0]
tiempo_futuro = ultimo_tiempo + 43200
precio_predicho = modelo.predict([[tiempo_futuro]])[0]

print(f"Precio predicho para 30 dÃ­as en el futuro (extrapolaciÃ³n lineal): ${precio_predicho:.2f}")
```

```python

# 5. Ajuste bayesiano usando el volumen
# Calcular el volumen promedio de los Ãºltimos 7 dÃ­as (7 * 24 * 60 = 10080 minutos)
ultimos_7_dias = df.tail(10080)
volumen_promedio = ultimos_7_dias['volume_(BTC)'].mean()
volumen_actual = ultimos_7_dias['volume_(BTC)'].iloc[-1]

# Calcular el cambio relativo en el volumen
cambio_volumen = volumen_actual / volumen_promedio

# Ajuste bayesiano simplificado: si el volumen aumenta mÃ¡s del 20%, asumimos un 5% de aumento en el precio
if cambio_volumen > 1.2:
    factor_ajuste = 1.05  # Aumento del 5%
else:
    factor_ajuste = 1.0  # Sin ajuste

precio_ajustado = precio_predicho * factor_ajuste
print(f"Precio ajustado (bayesiano): ${precio_ajustado:.2f}")
```
